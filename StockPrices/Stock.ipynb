# Stocks of multiple banks

```python
bklist =[
    'PNB.NS',
    'YESBANK.NS',
    'HDFCBANK.NS',
    'CANBK.NS',
    'MAHABANK.NS',
    'UJJIVANSFB.NS',
    'SBIN.NS',
    'KOTAKBANK.NS',
    'AXISBANK.BO',
    'IBN'
]
```

```python
histories = {}
```

```python
for s in bklist:
  tkr = yf.Ticker(s)
  print(tkr)
  print(tkr.info['financialCurrency'])
  bhist = tkr.history(period="5y")
  histories[s] = bhist
```

```python
import plotly.graph_objects as go
for stock in bklist:
  df = histories[stock].copy()
  fig = go.Figure(data=[go.Candlestick(x=df.index,
                  open=df['Open'], high=df['High'],
                  low=df['Low'], close=df['Close'])
                      ])

  fig.update_layout(margin= dict(l=15,r=15,t=50,b=20),
                    height= 500,
                    xaxis_rangeslider_visible=True,
                    title=f"{stock}")
  fig.show()
```
